```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.metrics import precision_score
import pandas as pd
```

```python
data = pd.read_csv("iris.csv")
target = data["Species"]
data = data.drop("Species", axis =1)

dataIono = pd.read_csv("ionosphere.csv", header = None)
targetIono = dataIono[dataIono.columns[34]]
dataIono = dataIono.drop(dataIono.columns[34], axis =1)

dataPima = pd.read_csv("pima-indians-diabetes.csv", header =None)
targetPima = dataPima[dataPima.columns[8]]
dataPima = dataPima.drop(dataPima.columns[8], axis =1)
```

```python
model = DecisionTreeClassifier(max_depth=3)

print("IRIS")
print()

for i in range(0, 15):
    X_train, X_test, y_train, y_test = train_test_split(data, target, test_size = 0.33, stratify = target)
    model = model.fit(X_train, y_train)
    a = tree.plot_tree(model)
    print(a)
    plt.show()
    print("Broj iteracije: ", i+1)
    print(model.get_n_leaves())
    print("Trening preciznost: ", precision_score(y_train, model.predict(X_train), average ="micro"))
    print("Test preciznost: ", precision_score(y_test, model.predict(X_test), average ="micro"))
    print()
    print()
    
print()
print()
print("IONOSPHERE")
print()
model = DecisionTreeClassifier(max_depth=3)

for i in range(0, 15):
    X_train, X_test, y_train, y_test = train_test_split(dataIono, targetIono, test_size = 0.33, stratify = targetIono)
    model = model.fit(X_train, y_train)
    a = tree.plot_tree(model)
    print(a)
    plt.show()
    print("Broj iteracije: ", i+1)
    print(model.get_n_leaves())
    print("Trening preciznost: ", precision_score(y_train, model.predict(X_train), pos_label = 'g'))
    print("Test preciznost: ", precision_score(y_test, model.predict(X_test), pos_label = 'g'))
    print()
    print()
    
print()
print()
print("PIMA")
print()
model = DecisionTreeClassifier(max_depth=5)
    
for i in range(0, 15):
    X_train, X_test, y_train, y_test = train_test_split(dataPima, targetPima, test_size = 0.33, stratify = targetPima)
    model = model.fit(X_train, y_train)
    a = tree.plot_tree(model)
    print(a)
    plt.show()
    print("Broj iteracije: ", i+1)
    print(model.get_n_leaves())
    print("Trening preciznost: ", precision_score(y_train, model.predict(X_train)))
    print("Test preciznost: ", precision_score(y_test, model.predict(X_test)))
    print()
    print()
```
