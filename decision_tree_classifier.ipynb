```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import precision_score
import pandas as pd
import numpy as np
```

```python
iris_data = pd.read_csv('iris.csv')
ionosphere_data = pd.read_csv('ionosphere.csv', header = None)
pimaIndiansDiabetes_data = pd.read_csv('pima-indians-diabetes.csv', header = None)
```

```python
iris_columnNames = iris_data.columns
iris_class = iris_data[iris_columnNames[4]]
iris_data = iris_data[iris_columnNames[:4]]

iris_trainData, iris_testData, iris_trainClass, iris_testClass = train_test_split(iris_data, iris_class, test_size = 0.33)
```

```python
ionosphere_class = ionosphere_data.iloc[:, 34:]
ionosphere_data = ionosphere_data.iloc[:, :34]

ionosphere_trainData, ionosphere_testData, ionosphere_trainClass, ionosphere_testClass = train_test_split(ionosphere_data, ionosphere_class, test_size = 0.33)
```

```python
pimaIndiansDiabetes_class = pimaIndiansDiabetes_data.iloc[:, 8:]
pimaIndiansDiabetes_data = pimaIndiansDiabetes_data.iloc[:, :8]

pimaIndiansDiabetes_trainData, pimaIndiansDiabetes_testData, pimaIndiansDiabetes_trainClass, pimaIndiansDiabetes_testClass = train_test_split(pimaIndiansDiabetes_data, pimaIndiansDiabetes_class, test_size = 0.33)
```

```python
model = DecisionTreeClassifier()
```

```python
model.fit(iris_trainData, iris_trainClass)
iris_predictedClass = model.predict(iris_testData)

print(precision_score(iris_testClass, iris_predictedClass, average='micro'))
```

```python
model.fit(ionosphere_trainData, ionosphere_trainClass)
ionosphere_predictedClass = model.predict(ionosphere_testData)

print(precision_score(ionosphere_testClass, ionosphere_predictedClass, average='micro'))
```

```python
model.fit(pimaIndiansDiabetes_trainData, pimaIndiansDiabetes_trainClass)
pimaIndiansDiabetes_predictedClass = model.predict(pimaIndiansDiabetes_testData)

print(precision_score(pimaIndiansDiabetes_testClass, pimaIndiansDiabetes_predictedClass, average='micro'))
```
